library "requests" is for connecting to webpage

```python
from bs4 import BeautifulSoup
import requests
```

### Get the csv example

Load in the html, request to visit the website, and get the entire webpage, store in "web_html"

```python
webpage_html = requests.get("https://simple.wikipedia.org/wiki/List_of_U.S._states_by_traditional_abbreviation")
soup = BeautifulSoup(webpage_html.text, 'html.parser')
soup
```

Get the csv table under the header "States"

```python
Ex_section = soup.find(id='States')     #locate where the section is
states_table = Ex_section.findNext('table')  #locate the immediate next <pre> tag
states_table
```

### Save the HTML of the "States" table to a file


```python
with open('states.html', 'w', encoding='utf-8') as f:
    f.write(str(states_table))
```

Reload the csv data from the file to make sure the data was saved properly

Read the HTML table using pandas

```python
import pandas as pd
tables = pd.read_html('states.html')
```

Assuming the first table is the one you want

```python
states_data = tables[0]
```

Extract only essential columns

```python
selected_columns = states_data.iloc[:, [0, 2]]
```

Save the csv example into a csv file

```python
selected_columns.to_csv('selected_columns.csv', index=False)
```
